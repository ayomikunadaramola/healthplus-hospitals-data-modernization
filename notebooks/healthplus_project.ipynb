### Clean Age

```python
patients["age"]
```

```python
# Converting the age values to integers/numeric 
patients["age"] = pd.to_numeric(
    patients["age"], 
    errors="coerce" #Replace anything that cannot be converted to a number with NaN(Not a Number). Invalid age become NaN
)
```

```python
# Replace missing age values 
patients["age"] = patients["age"].fillna(
    patients["age"].median()
)
```

```python
patients["age"] # Here, we are still getting float values and negative values. We will convert the age values to integers and replace negative values with the median age value.
```

```python
# Replace negative ages with missing values
patients.loc[patients["age"] < 0, "age"] = pd.NA # what this line of code does is that it replaces any negative age values with a missing value (NaN). This is done because negative ages are not valid and should be treated as missing data. By replacing them with NaN, we can then handle them appropriately, such as filling them with the median age value later on. 

# Fill missing values with the median
patients["age"] = patients["age"].fillna(
    patients["age"].median()
)

# Round ages
patients["age"] = patients["age"].round()

# Convert to integers
patients["age"] = patients["age"].astype(int)
```

- Every record must have an ID.
- Every ID must be unique.
- IDs cannot be missing.

If an ID is missing, you have three possible approaches:

- Retrieve the correct ID from the source system (best in production).
- Generate a new unique ID (acceptable for this teaching project).
- Remove the record if it cannot be trusted (appropriate in some real-world scenarios).

### Remove spaces in the phone_number column 

```python
patients["phone_number"] = patients["phone_number"].str.replace(" ", "", regex=False) # this means that any spaces in the phone_number column will be removed. The regex=False argument indicates that we are not using regular expressions for this replacement, but rather a simple string replacement. This is done to ensure that the phone numbers are in a consistent format without any spaces.
```

### Remove Hyphens in the phone_number column 

```python
patients["phone_number"] = patients["phone_number"].str.replace("-", "", regex=False) # this means that any hyphens in the phone_number column will be removed. The regex=False argument indicates that we are not using regular expressions for this replacement, but rather a simple string replacement. This is done to ensure that the phone numbers are in a consistent format without any hyphens.
```

### Convert +234 numbers to Nigerian format

```python
patients["phone_number"] = patients["phone_number"].str.replace(
    r"^\+234",
    "0",
    regex=True
)
```

```python
patients["phone_number"]
```

### Add the missing leading zero 


```python
patients["phone_number"] = patients["phone_number"].apply(
    lambda x: "0" + x
    if isinstance(x, str) and x.isdigit() and len(x) == 10
    else x
)
```

```python
patients["phone_number"]
```

```python
patients["phone_number"] = patients["phone_number"].replace({
    "not_available": "Unknown"
})
```

```python
patients["phone_number"]
```

```python
# Replace invalid phone numbers
# A Nigerian phone number should contain 11 digits.
#Anything shorter or longer is invalid. 

patients["phone_number"] = patients["phone_number"].apply(
    lambda x: "Unknown"
    if isinstance(x, str)
    and x != "Unknown"
    and (not x.isdigit() or len(x) != 11)
    else x
)

"""
This will automatically convert:
12345 into Unknown

while leaving 08026951234 unchanged.
"""
```

```python
patients["phone_number"]
```

### Clean the appointment_id missing rows/values 
Since appointment_id is the primary key of the appointments table, you should not fill missing values with "Unknown" or leave them blank.

A primary key must satisfy three rules:

- Unique
- Not Null
- Stable (should not change)

```python
# Find the largest appointment_id
# Get the highest appointment number
max_id = (
    appointments["appointment_id"]
    .dropna()
    .str.replace("A", "", regex=False)
    .astype(int)
    .max()
)

print(max_id)
```

```python
# Generate new appointment IDs for missing values 
# Find rows with missing IDs
missing_rows = appointments["appointment_id"].isna()

# Create new IDs
new_ids = [
    f"A{num:06d}"
    for num in range(max_id + 1, max_id + 1 + missing_rows.sum())
]

# Assign them
appointments.loc[missing_rows, "appointment_id"] = new_ids
```
